```python
# import the necessary packages
from pytesseract import Output
import matplotlib.pyplot as plt
import pytesseract
import argparse
import imutils
import cv2
```

```python
def plt_imshow(title, image):
	# convert the image frame BGR to RGB color space and display it
	image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
	plt.imshow(image)
	plt.title(title)
	plt.grid(False)
	plt.show()
```

```python

args = {
	"image_normal": "/content/normal.jpg",
	"image_rotated_180": "/content/helpdesk_180.png",
	"image_rotated_90_clockwise": "/content/rotated_270.png",
	"image_rotated_90_counter_clockwise": "/content/rotated_270.jpg",
}
```

```python
# load the input image, convert it from BGR to RGB channel ordering,
# and use Tesseract to determine the text orientation
image = cv2.imread(args["image_normal"])
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
results = pytesseract.image_to_osd(rgb, output_type=Output.DICT)

# display the orientation information
print("[INFO] detected orientation: {}".format(
	results["orientation"]))
print("[INFO] rotate by {} degrees to correct".format(
	results["rotate"]))
print("[INFO] detected script: {}".format(results["script"]))

# rotate the image to correct the orientation
rotated = imutils.rotate_bound(image, angle=results["rotate"])

# show the original image and output image after orientation
# correction
plt_imshow("Original", image)
plt_imshow("Output", rotated)
```
